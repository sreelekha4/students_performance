```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students_scores.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
plt.title("gender distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analyzed that:
#the number of females in the data is more than the number of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("relationship between parent's education and student's score")
plt.show()
```

```python
#from the above chart we conclude that the education of the parents have a good impact on their students marks
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("relationship between parent marital status and student's score")
plt.show()
```

```python
#from the above chart we have concluded that there is no impact on the
#student's score due to their parent's marital status
```
